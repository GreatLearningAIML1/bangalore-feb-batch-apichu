```python
import pandas as pd
df =pd.read_csv('data.csv') 
```

```python
import numpy as np
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro
from statsmodels.stats.power import ttest_power
```

```python

group1 = df["Old Scheme (in thousands)"]
group2 = df["New Scheme (in thousands)"]
```

```python
#Question 1 : Mean

Old_scheme = group1.mean();
print("Mean of Old Scheme ( in thousands)" , round(Old_scheme,2))
```

```python
#Question 1 : Mean
New_scheme = group2.mean();
print("Mean of New Scheme ( in thousands)" , round(New_scheme,2))
```

```python
t_statistic, p_value = ttest_ind(group1, group2)
#Null hypothesis:  New scheme has not significantly raised compare to Old Scheme
#Alternative hypothesis: New scheme has significantly raised compare to Old Scheme

```

```python
#Question 2 : P value
print("The p Vale = :", round(p_value,2) )
```

```python
#Question 3 : p-value
# p-vale is >0.05 Hence . Null Hypothesis not Rejected New scheme not raised output significatly 
```

```python
#Question 4 :
#Suppose it has been calculated that in order for Titan to break even, 
#the average output must increase by £5000 in the scheme compared to the old scheme.
#If this figure is alternative hypothesis, what is:

#Null hypothesis:  old scheme + £5000 has not breakeven for  Titan compare to old schame
#Alternative hypothesis: old scheme + £5000 has breakeven for  Titan compare to old schame
t_statistic, p_value = ttest_ind(group1,group1+5)
print("P-Value :",round(p_value,3))
```

```python
#Question 4 (a) :
#The probability of a type 1 error? (2.5 points)
#The Calcualted p-value of 0.3477 is the probability of committing a Type I error ( Chance of getting it wrong). 
#A p-Value 0.35 is a high probability of making a mistake , so we can not concluded that old scheme + £5000 has 
# breakeven for Titan compare to old schame
```

```python
#Question 4 (b) : What is the p- value of the hypothesis test if we test for a difference of $5000? (5 points)
# $5000 = £3831. 
t_statistic, p_value = ttest_ind(group1,group1+3.831)
print("p-value of the hypothesis test if we test for a difference of $5000 is equal to ", round(p_value,2))
```
